```python
import matplotlib.pyplot as plt
```

### Line Chart

```python
days=["monday","tuesday","wednesday","thursday","friday","saturday"]
sales=[200,450,300,500,650,700]
plt.plot(days,sales,marker="o")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.grid(True)
plt.title("Daily Sales")
plt.show()
```

### Bar Chart

```python
product=['A','B','C','D','E']
revenue=[100,200,300,400,326]
plt.bar(product,revenue,color="skyblue")
plt.xlabel("product")
plt.ylabel("revenue")
plt.show()
```

### Pie Chart

```python
label=["iOS","Android","Others"]
market_share=[40,52,63]
plt.pie(market_share,labels=label ,startangle=140,autopct='%1.1f%%')
plt.title("Mobile os market share")
plt.axis('equal') # equal aspect ratio for perfect circle
plt.show()
```

### Histogram

```python
ages=[22,34,23,12,56,25,45,30,32,54,61,42,32]
plt.hist(ages,bins=5,color='indigo',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```

### Scatter Plot

```python
income=[30,35,40,45,50,55]
spending=[20,25,30,35,40,45]
plt.scatter(income,spending,color="red")
plt.title("Income vs Spending")
plt.xlabel("Income(k)")
plt.ylabel("Spending(k)")
plt.show()
```

### Heatmap - Correlation matrix (with Seaborn)

```python
import seaborn as sns
import pandas as pd
```

```python
data={
    'Math':[90,80,85,95],
    'Science':[85,89,92,880,],
    'English':[78,75,80,85]
}

df=pd.DataFrame(data)
correlation=df.corr()
sns.heatmap(correlation,cmap='coolwarm',annot=True)
plt.title("Subject score correlation")
plt.show()
```
